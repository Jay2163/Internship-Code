```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Desktop/laptop_data.csv")
```

```python
df.head()
```

# 1.Cateogrical Data

# a.Countplot

```python
plot = sns.countplot(df["Ram"])
plot.set_xticklabels(plot.get_xticklabels(), rotation=40);
```

# b.pichart

```python
df["TypeName"].value_counts().plot(kind="pie",autopct="%0.2f")
```

# 2.Numerical Data

# a.Histogram

```python
plt.hist(df["Inches"],bins=10)
```

# b.Distplot

```python
sns.distplot(df["Inches"])
```

# c.boxplot

```python
sns.boxplot(df["Inches"]) #use for noisy data
```
